```python
edmonton_crimes_data_path_2022 = "../../data/01_raw_data/05_crime_by_city_data/Edmonton/Crimes_2022.csv"
edmonton_crimes_data_path_2023 = "../../data/01_raw_data/05_crime_by_city_data/Edmonton/Crimes_2023.csv"
edmonton_crimes_data_path_2024 = "../../data/01_raw_data/05_crime_by_city_data/Edmonton/Crimes_2024.csv"
edmonton_crimes_data_path_2025 = "../../data/01_raw_data/05_crime_by_city_data/Edmonton/Crimes_2025.csv"

edmonton_crimes = pd.concat([
    pd.read_csv(edmonton_crimes_data_path_2022),
    pd.read_csv(edmonton_crimes_data_path_2023),
    pd.read_csv(edmonton_crimes_data_path_2024),
    pd.read_csv(edmonton_crimes_data_path_2025),
], ignore_index=True)
```

```python
from pyproj import Transformer

source_crs = "EPSG:3857" 
target_crs = "EPSG:4326"  

# 1. Create transformer
transformer = Transformer.from_crs(source_crs, target_crs, always_xy=True)

# 2. Extract the columns
x_series = edmonton_crimes['x']
y_series = edmonton_crimes['y']

# 3. Apply the transformation to the entire columns at once (vectorized!)
# The output is a tuple (longitude, latitude)
longitude, latitude = transformer.transform(x_series, y_series)

# 4. Update the DataFrame with the new columns
edmonton_crimes['longitude'] = longitude
edmonton_crimes['latitude'] = latitude

# Optional: Drop the old x and y columns if no longer needed
edmonton_crimes = edmonton_crimes.drop(columns=['x', 'y'])

# Display the result to confirm (e.g., the first few rows)
print(edmonton_crimes.head())
```

#### data is inaccurate for x and y, so we will use Google Maps API to get lat/long from intersection info

```python
import requests


GOOGLE_API_KEY = "GOOGLE_API_KEY"

def get_lat_long(address):
    """
    Calls Google Geocoding API and returns (lat, lng)
    """
    url = "https://maps.googleapis.com/maps/api/geocode/json"
    params = {"address": address, "key": GOOGLE_API_KEY}

    response = requests.get(url, params=params)
    data = response.json()

    if data["status"] == "OK":
        location = data["results"][0]["geometry"]["location"]
        return location["lat"], location["lng"]

    return None, None
```

```python
edmonton_crimes['Full_Address'] = edmonton_crimes['Intersection'] + ", Edmonton, AB, Canada"
```

```python
edmonton_crimes
```

```python
unique_addresses = edmonton_crimes['Full_Address'].dropna().unique()
print(f"Total rows: {len(edmonton_crimes)}")
print(f"Unique addresses to process: {len(unique_addresses)}")
```

```python
coordinate_map = {}

for i, address in enumerate(unique_addresses):
    # Call the geocoding function
    lat, lng = get_lat_long(address)
    
    print(f"Processing {i+1}/{len(unique_addresses)}: {address} - {lat}, {lng}")
   
    # Store the result in the map
    coordinate_map[address] = (lat, lng)

print("Geocoding complete.")
```

```python
edmonton_crimes['coordinates'] = edmonton_crimes['Full_Address'].map(coordinate_map)

edmonton_crimes[['latitude', 'longitude']] = pd.DataFrame(
    edmonton_crimes['coordinates'].tolist(), 
    index=edmonton_crimes.index
)

# Clean up by dropping the temporary 'coordinates' column
df = edmonton_crimes.drop(columns=['coordinates'])

# Display the result
print("\n--- Final DataFrame Head ---")
print(edmonton_crimes.head())
```

```python
edmonton_crimes = edmonton_crimes.drop(columns=['coordinates'])
edmonton_crimes
```

```python
from pyproj import Transformer

# NAD83 / UTM Zone 10N is commonly represented by EPSG:26910
source_crs = "EPSG:26910" 
target_crs = "EPSG:4326"  # WGS 84 (standard Lat/Lon)

# 1. Create transformer
# always_xy=True ensures that the input is (Easting, Northing) and output is (Longitude, Latitude)
transformer = Transformer.from_crs(source_crs, target_crs, always_xy=True)

# 2. Extract the columns (The given UTM coordinates)
x_series = 490247.1512
y_series = 5459171.0415

# 3. Apply the transformation
longitude, latitude = transformer.transform(x_series, y_series)

# The coordinates will be the WGS 84 result (approx. 49Â°N, 122Â°W)
print(f"Latitude: {latitude}, Longitude: {longitude}")
print(f"{latitude},{longitude}")
```
