```python
import pandas as pd
```

```python
sales_df = pd.read_csv("../../data/02_processed_data/02_cannabis_sales/sales_data.csv")
```

```python
unique_data_location = sales_df["DGUID"].unique()
```

```python
crime_df = pd.read_parquet('../../data/01_raw_data/04_crime_data/crime_data.parquet')
```

```python
crime_df.info()
```

```python
crime_df["GEO"].nunique()
```

```python
crime_df["GEO"].unique()
```

```python
new_df = crime_df.copy()
new_df['Violation Name'] = crime_df['Violations'].str.replace(r'\s*\[[^\]]+\]$', '', regex=True).str.strip()
new_df['Violation Code'] = crime_df['Violations'].str.extract(r'\[([^\]]+)\]').squeeze()
```

```python
new_df.head(10)
```

```python
provice_and_canada_df = new_df[new_df["DGUID"].isin(unique_data_location)]
```

```python
provice_and_canada_df.shape
```

```python
provice_and_canada_df.head()
```

```python
selected_cols = ['REF_DATE', 'GEO','DGUID', 'Violation Name', 'Violation Code', 'Statistics', 'UOM', 'VALUE']

viz_df = provice_and_canada_df[selected_cols].copy()

viz_df = viz_df.rename(columns={
    'REF_DATE': 'year',
    'Location Name': 'location',
    'Violation Name': 'violation',
    'VALUE': 'value'
})

viz_df = viz_df.dropna(subset=['value']).reset_index(drop=True)

viz_df.head()
```

```python
viz_df.to_parquet('../../data/02_processed_data/04_crime_data/canada_and_province_crime_data.parquet', index=False)
```

```python
provice_and_canada_df[["GEO", "DGUID"]].drop_duplicates()
```

```python
city_df = new_df[~new_df["DGUID"].isin(unique_data_location)][selected_cols].drop_duplicates().dropna(subset=['DGUID'])
```

```python
city_df.to_parquet('../../data/02_processed_data/04_crime_data/city_wise_crime_data.parquet', index=False)
```
